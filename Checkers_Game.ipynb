Imports

```python
import pandas as pd
```

Defines

```python
BLACK = 'B'
WHITE = 'W'
BLACK_KING = 'BK'
WHITE_KING = 'WK'

#For Customizing Game with Bigger Board Change Following Parameters
BOARD_SIZE_XY = 8
TOTAL_CHECKERS = 24
KING_ROW_WHITE = 7
KING_ROW_BLACK = 0
## Till Here 

FORWORD_LEFT = 1
FORWORD_RIGHT = 2
BACKWORD_LEFT = 3
BACKWORD_RIGHT = 4
FORWORD_LEFT_JUMP = 5
FORWORD_RIGHT_JUMP = 6
BACKWORD_LEFT_JUMP = 7
BACKWORD_RIGHT_JUMP = 8


NORMAL_MOVE = 1
JUMP_MOVE = 2
NO_MOVE = 0

EMPTY_BLOCK = -1
BOUND_ERROR = -2
```

Checker Class and its Methords

```python
class Checkers:
    Player = BLACK 
    Pos_x = 0 
    Pos_y = 0
    King = False
    Alive = True
    
    def __init__(self, Player, x,y):
        self.Player = Player
        self.Pos_x = x
        self.Pos_y = y
        self.King = False
        self.Alive = True
        self.ForcedCapture = False
```

Checkers Declarations and Initilizaton

```python
# To be Modified as Per Custom Board Size

Checker = []

A1 = Checkers(BLACK, 0, 1)
A2 = Checkers(BLACK, 0, 3)
A3 = Checkers(BLACK, 0, 5)
A4 = Checkers(BLACK, 0, 7)
A5 = Checkers(BLACK, 1, 0)
A6 = Checkers(BLACK, 1, 2)
A7 = Checkers(BLACK, 1, 4)
A8 = Checkers(BLACK, 1, 6)
A9 = Checkers(BLACK, 2, 1)
A10 = Checkers(BLACK,2, 3)
A11 = Checkers(BLACK,2, 5)
A12 = Checkers(BLACK,2, 7)


B12 = Checkers(WHITE,5, 0)
B11 = Checkers(WHITE,5, 2)
B10 = Checkers(WHITE,5, 4)
B9 = Checkers(WHITE, 5, 6)
B8 = Checkers(WHITE, 6, 1)
B7 = Checkers(WHITE, 6, 3)
B6 = Checkers(WHITE, 6, 5)
B5 = Checkers(WHITE, 6, 7)
B4 = Checkers(WHITE, 7, 0)
B3 = Checkers(WHITE, 7, 2)
B2 = Checkers(WHITE, 7, 4)
B1 = Checkers(WHITE, 7, 6)

Checker = [A1,A2,A3,A4,A5,A6,A7,A8,A9,A10,A11,A12,B1,B2,B3,B4,B5,B6,B7,B8,B9,B10,B11,B12]
```

Game Methords Globals

```python
playerBlack = 'BLACK'
playerWhite = 'WHITE'

Turn = BLACK #Intial Turn
update_x = -1
update_y = -1
jump_checkerIndex = -1
updatedCheckerIndex = -1

ForcedCaptures = False
```

Game Methords

```python
def getUpdated_x():
    return Checker[updatedCheckerIndex].Pos_x

def getUpdated_y():
    return Checker[updatedCheckerIndex].Pos_y
    

def getCheckerIndex(x,y,Player):
    if not (x >= 0 and x < BOARD_SIZE_XY and y >= 0 and y < BOARD_SIZE_XY):
        return BOUND_ERROR
    
    if Player is None:
        for i in range(TOTAL_CHECKERS):
            if (Checker[i].Alive and Checker[i].Pos_x == x and Checker[i].Pos_y == y):
                return i
                break
    else:
        for i in range(TOTAL_CHECKERS):
            if (Checker[i].Alive and Checker[i].Pos_x == x and Checker[i].Pos_y == y and Checker[i].Player == Player): 
                return i
                break
    return EMPTY_BLOCK


def switchTurn():
    global Turn
    if Turn == WHITE:
        Turn = BLACK
    else:
        Turn = WHITE


def showBoard():
    game = [[' ' for x in range(BOARD_SIZE_XY)] for y in range(BOARD_SIZE_XY)]
    for i in range(TOTAL_CHECKERS):
        if Checker[i].Alive and Checker[i].Player == BLACK:
            if Checker[i].King:
                game[Checker[i].Pos_x][Checker[i].Pos_y] = BLACK_KING
            else:
                game[Checker[i].Pos_x][Checker[i].Pos_y] = BLACK
        if Checker[i].Alive and Checker[i].Player == WHITE:
            if Checker[i].King:
                game[Checker[i].Pos_x][Checker[i].Pos_y] = WHITE_KING
            else:
                game[Checker[i].Pos_x][Checker[i].Pos_y] = WHITE
    print "\nR _________BLACK________ L\n"
    print pd.DataFrame(game)
    print "\nL _________WHITE________ R\n"
    
def input_xyValid(x,y):
    global Turn
    for i in range(TOTAL_CHECKERS):
            if Checker[i].Alive and Checker[i].Pos_x == x and Checker[i].Pos_y == y:
                if Turn == Checker[i].Player:
                    return True
                    break
    return False
```

Game Rules Methords

```python
#All can_____ Methords Updates Global Variables 
#To be taken Care while Making makeMove(...) or makeJump(...)
    
def canMoveForwordLeft(x,y):
    global update_x
    global update_y
    checkerIndex = getCheckerIndex(x,y, None)
    
    if Checker[checkerIndex].Player == BLACK:
        update_x = Checker[checkerIndex].Pos_x + 1
        update_y = Checker[checkerIndex].Pos_y + 1
    else:
        update_x = Checker[checkerIndex].Pos_x - 1
        update_y = Checker[checkerIndex].Pos_y - 1

    if getCheckerIndex(update_x, update_y, None) == EMPTY_BLOCK:
        return True
    return False
    
def canMoveForwordRight(x,y):
    global update_x
    global update_y
    checkerIndex = getCheckerIndex(x,y, None)
    if Checker[checkerIndex].Player == BLACK:
        update_x = Checker[checkerIndex].Pos_x + 1
        update_y = Checker[checkerIndex].Pos_y - 1
    else:
        update_x = Checker[checkerIndex].Pos_x - 1
        update_y = Checker[checkerIndex].Pos_y + 1
    if getCheckerIndex(update_x, update_y, None) == EMPTY_BLOCK:
        return True
    return False

def canMoveBackwordLeft(x,y):
    global update_x
    global update_y
    checkerIndex = getCheckerIndex(x,y, None)
    if Checker[checkerIndex].King:
        if Checker[checkerIndex].Player == BLACK:
            update_x = Checker[checkerIndex].Pos_x - 1
            update_y = Checker[checkerIndex].Pos_y + 1
        else:
            update_x = Checker[checkerIndex].Pos_x + 1
            update_y = Checker[checkerIndex].Pos_y - 1
        if getCheckerIndex(update_x, update_y, None) == EMPTY_BLOCK:
            return True
    return False

def canMoveBackwordRight(x,y):
    global update_x
    global update_y
    checkerIndex = getCheckerIndex(x,y, None)
    if Checker[checkerIndex].King:
        if Checker[checkerIndex].Player == BLACK:
            update_x = Checker[checkerIndex].Pos_x - 1
            update_y = Checker[checkerIndex].Pos_y - 1
        else:
            update_x = Checker[checkerIndex].Pos_x + 1
            update_y = Checker[checkerIndex].Pos_y + 1

        if getCheckerIndex(update_x, update_y, None) == EMPTY_BLOCK:
            return True
    return False

def canJumpForwordLeft(x,y):
    global Turn
    global update_x
    global update_y
    global jump_checkerIndex
    checkerIndex = getCheckerIndex(x,y, None)
    if Checker[checkerIndex].Player == BLACK:
        update_x = Checker[checkerIndex].Pos_x + 2
        update_y = Checker[checkerIndex].Pos_y + 2
        jump_x = Checker[checkerIndex].Pos_x + 1
        jump_y = Checker[checkerIndex].Pos_y + 1
    else:
        update_x = Checker[checkerIndex].Pos_x - 2
        update_y = Checker[checkerIndex].Pos_y - 2
        jump_x = Checker[checkerIndex].Pos_x - 1
        jump_y = Checker[checkerIndex].Pos_y - 1
    
    if getCheckerIndex(update_x, update_y, None) == EMPTY_BLOCK:
        jump_checkerIndex = getCheckerIndex(jump_x, jump_y, None)  ##Checking Opposite Player is Available
        if jump_checkerIndex != EMPTY_BLOCK and jump_checkerIndex != BOUND_ERROR:
            if Checker[jump_checkerIndex].Player != Turn:
                return True
    return False
    
def canJumpForwordRight(x,y):
    global Turn
    global update_x
    global update_y
    global jump_checkerIndex
    checkerIndex = getCheckerIndex(x,y, None)
    if Checker[checkerIndex].Player == BLACK:
        update_x = Checker[checkerIndex].Pos_x + 2
        update_y = Checker[checkerIndex].Pos_y - 2
        jump_x = Checker[checkerIndex].Pos_x + 1
        jump_y = Checker[checkerIndex].Pos_y - 1
    else:
        update_x = Checker[checkerIndex].Pos_x - 2
        update_y = Checker[checkerIndex].Pos_y + 2
        jump_x = Checker[checkerIndex].Pos_x - 1
        jump_y = Checker[checkerIndex].Pos_y + 1
    
    if getCheckerIndex(update_x, update_y, None) == EMPTY_BLOCK:
        jump_checkerIndex = getCheckerIndex(jump_x, jump_y, None)  ##Checking Opposite Player is Available
        if jump_checkerIndex != EMPTY_BLOCK and jump_checkerIndex != BOUND_ERROR:
            if Checker[jump_checkerIndex].Player != Turn:
                return True
    return False

def canJumpBackwordLeft(x,y):
    global Turn
    global update_x
    global update_y
    global jump_checkerIndex
    checkerIndex = getCheckerIndex(x,y, None)
    if Checker[checkerIndex].King:
        if Checker[checkerIndex].Player == BLACK:
            update_x = Checker[checkerIndex].Pos_x - 2
            update_y = Checker[checkerIndex].Pos_y + 2
            jump_x = Checker[checkerIndex].Pos_x - 1
            jump_y = Checker[checkerIndex].Pos_y + 1
        else:
            update_x = Checker[checkerIndex].Pos_x + 2
            update_y = Checker[checkerIndex].Pos_y - 2
            jump_x = Checker[checkerIndex].Pos_x + 1
            jump_y = Checker[checkerIndex].Pos_y - 1

        if getCheckerIndex(update_x, update_y, None) == EMPTY_BLOCK:
            jump_checkerIndex = getCheckerIndex(jump_x, jump_y, None)  ##Checking Opposite Player is Available
            if jump_checkerIndex != EMPTY_BLOCK and jump_checkerIndex != BOUND_ERROR:
                if Checker[jump_checkerIndex].Player != Turn:
                    return True
    return False

def canJumpBackwordRight(x,y):
    global Turn
    global update_x
    global update_y
    global jump_checkerIndex

    checkerIndex = getCheckerIndex(x,y, None)
    if Checker[checkerIndex].King:
        if Checker[checkerIndex].Player == BLACK:
            update_x = Checker[checkerIndex].Pos_x - 2
            update_y = Checker[checkerIndex].Pos_y - 2
            jump_x = Checker[checkerIndex].Pos_x - 1
            jump_y = Checker[checkerIndex].Pos_y - 1
        else:
            update_x = Checker[checkerIndex].Pos_x + 2
            update_y = Checker[checkerIndex].Pos_y + 2
            jump_x = Checker[checkerIndex].Pos_x + 1
            jump_y = Checker[checkerIndex].Pos_y + 1

        if getCheckerIndex(update_x, update_y, None) == EMPTY_BLOCK:
            jump_checkerIndex = getCheckerIndex(jump_x, jump_y, None)  ##Checking Opposite Player is Available
            if jump_checkerIndex != EMPTY_BLOCK and jump_checkerIndex != BOUND_ERROR:
                if Checker[jump_checkerIndex].Player != Turn:
                    return True
    return False

def makeMove(x,y,move):
    global updatedCheckerIndex
    #to Update Proper Paramenters
    if move == FORWORD_LEFT:
        canMoveForwordLeft(x,y)
    elif move == FORWORD_RIGHT:
        canMoveForwordRight(x,y)
    elif move == BACKWORD_LEFT:
        canMoveBackwordLeft(x,y)
    elif move == BACKWORD_RIGHT:
        canMoveBackwordRight(x,y)
    else:
        return False
    checkerIndex = getCheckerIndex(x,y, None)
    Checker[checkerIndex].Pos_x = update_x
    Checker[checkerIndex].Pos_y = update_y
    updatedCheckerIndex = checkerIndex
    return True
    
def makeJump(x,y,jump):
    global updatedCheckerIndex
    #Update Proper Global Paramenters as per Move
    if jump == FORWORD_LEFT_JUMP:
        canJumpForwordLeft(x,y)
    elif jump == FORWORD_RIGHT_JUMP:
        canJumpForwordRight(x,y)
    elif jump == BACKWORD_LEFT_JUMP:
        canJumpBackwordLeft(x,y)
    elif jump == BACKWORD_RIGHT_JUMP:
        canJumpBackwordRight(x,y)
    else:
        return False
    
    Checker[jump_checkerIndex].Alive = False            
    
    checkerIndex = getCheckerIndex(x,y, None)
    Checker[checkerIndex].Pos_x = update_x
    Checker[checkerIndex].Pos_y = update_y
    Checker[checkerIndex].ForcedCapture = False
    updatedCheckerIndex = checkerIndex
    return True

def becameKing():
    checkerIndex = getCheckerIndex(getUpdated_x(),getUpdated_y(),None)
    if Checker[checkerIndex].Player == BLACK:
        if Checker[checkerIndex].Pos_x == KING_ROW_WHITE:
            Checker[checkerIndex].King = True
            return True
    else:
        if Checker[checkerIndex].Pos_x == KING_ROW_BLACK:
            Checker[checkerIndex].King = True
            return True
    return False
 
def canReJump():
    canJump = False
    x = getUpdated_x()
    y = getUpdated_y()
    if canJumpForwordLeft(x,y):
        canJump = True
    if canJumpForwordRight(x,y):
        canJump = True
    if canJumpBackwordLeft(x,y):
        canJump = True
    if canJumpBackwordRight(x,y):
        canJump = True
    return canJump
        
def canForcedCaptures():
    global Turn
    global ForcedCaptures
    
    canForceCapture = False
    for i in range(TOTAL_CHECKERS):
        if Checker[i].Alive and Checker[i].Player == Turn and ForcedCaptures:
            x = Checker[i].Pos_x
            y = Checker[i].Pos_y
            
            if canJumpForwordLeft(x,y):
                Checker[i].ForcedCapture = True
                canForceCapture = True
            if canJumpForwordRight(x,y):
                Checker[i].ForcedCapture = True
                canForceCapture = True
            if canJumpBackwordLeft(x,y):
                Checker[i].ForcedCapture = True
                canForceCapture = True
            if canJumpBackwordRight(x,y):
                Checker[i].ForcedCapture = True
                canForceCapture = True

    return canForceCapture            

def gameOver():
    global Turn
    canMakeMove = False
    for i in range(TOTAL_CHECKERS):
        if Checker[i].Alive and Checker[i].Player == Turn:
            x = Checker[i].Pos_x
            y = Checker[i].Pos_y
            
            if canMoveForwordLeft(x,y):
                canMakeMove = True
            if canMoveForwordRight(x,y):
                canMakeMove = True
            if canMoveBackwordLeft(x,y):
                canMakeMove = True
            if canMoveBackwordRight(x,y):
                canMakeMove = True
            if canJumpForwordLeft(x,y):
                canMakeMove = True
            if canJumpForwordRight(x,y):
                canMakeMove = True
            if canJumpBackwordLeft(x,y):
                canMakeMove = True
            if canJumpBackwordRight(x,y):
                canMakeMove = True
        if canMakeMove:
            return False
    return True            
```

Exception Handling Methords (Safe Calls)

```python
def isValidMoveInput(inp,move):
    if move >= 0 and move <= 9:
        try:
            isValidMoveInput = inp.index(move)
            return True
        except ValueError:
            isValidMoveInput = -1
            return False

def safe_input(inputLabel):
    validInput = False
    while(not validInput):
        try:
            inp = input(inputLabel)
            return inp
        except:
            print "# You didn't Entered any Input."
            
def safe_raw_input(inputLabel):
    validInput = False
    while(not validInput):
        try:
            inp = raw_input(inputLabel)
            return inp
        except:
            print "# You didn't Entered any Input."
```

Game Console Methords IO

```python
def move(moveType):
    global Turn
    global playerBlack
    global playerWhite
    global ForcedCaptures
    
    forcedCaptures = False
    
    canMove = False
    canJump = False
    
    if gameOver():
        print "######## GAME OVER ########\n\n"
        if Turn == BLACK:
            print "         WHITE WON         \n"
        else:
            print "         BLACK WON          \n"
        
        return NO_MOVE
        
    if canForcedCaptures():
        forcedCaptures = True
        
    #print "\n--------------------------------------------------------"
    if Turn == BLACK:
        print "::# " + playerBlack + "'s Turn :: B #::"
    if Turn == WHITE:
        print "::# " + playerWhite + "'s Turn :: W #::"
    
    if moveType == JUMP_MOVE:
        x = getUpdated_x()
        y = getUpdated_y()
        print "\nYou Still can Make Jumps"
    else:
        x = safe_input("Your Move at Y : ")
        y = safe_input("Your Move at X : ")
    
    inp = []
    if input_xyValid(x,y):
        print "\nPossible Moves::"
        if moveType == NORMAL_MOVE and not forcedCaptures:
            if canMoveForwordLeft(x,y):
                print "1. Move Forword Left"
                canMove = True
                inp.append(1)
            if canMoveForwordRight(x,y):
                print "2. Move Forword Right"
                canMove = True
                inp.append(2)
            if canMoveBackwordLeft(x,y):
                print "3. Move Backword Left"
                canMove = True
                inp.append(3)
            if canMoveBackwordRight(x,y):
                print "4. Move Backword Right"
                canMove = True
                inp.append(4)
        if canJumpForwordLeft(x,y):
            print "5. Jump Forword Left"
            canJump = True
            inp.append(5)
        if canJumpForwordRight(x,y):
            print "6. Jump Forword Right"
            canJump = True
            inp.append(6)
        if canJumpBackwordLeft(x,y):
            print "7. Jump Backword Left"
            canJump = True
            inp.append(7)
        if canJumpBackwordRight(x,y):
            print "8. Jump Backword Right"
            canJump = True
            inp.append(8)
        if not (canMove or canJump):
            print "-- None --"
            if forcedCaptures:
                print "There are Forced Captures in Game"
        if moveType == JUMP_MOVE and not forcedCaptures:
            print "9. End Move"
        print "0. To Reenter"
        inp.append(0)
        move = safe_input("Your Move : ")
        
        
        if isValidMoveInput(inp,move):
            if move == 0:
                return moveType
            elif move == 9:
                switchTurn()
                return NORMAL_MOVE
            elif move > 0 and move <=4:
                makeMove(x,y, move)
                becameKing()
                switchTurn()
            else:
                makeJump(x,y,move)
                becameKing()
                if canReJump():
                    return JUMP_MOVE
                else:
                    switchTurn()
        else:
            print "\n# That Move is Not Possible or Available."
    else:
        print "\n# Not your Turn or Invalid Block Selected."
    return NORMAL_MOVE
            

def startGame():
    global playerBlack
    global playerWhite
    global ForcedCaptures
    moveType = NORMAL_MOVE
    
    print "Welcome to Play Checkers. \n"
    print "You want to Play\n1.Forced Captures Game\n2.Non-Forced Captures Game"
    captureType = safe_input("Select :")
    if captureType == 1:
        ForcedCaptures = True
        print "Great! Game Mode is Set to Forced Captures."
    else:
        ForcedCaptures = False
        print "Cool! Game Mode is Set to Non - Forced Captures."
        
    print "\nPlease Make a Toss to Select Player BLACK.\n"
    playerBlack = safe_raw_input("Player Black Name : ")
    playerWhite = safe_raw_input("Player White Name : ")
    
    print "\n\nGreat! Lets Start..."
    
    showBoard()       
    while(1):
        if moveType == NORMAL_MOVE:
            moveType = move(NORMAL_MOVE)
            showBoard()
        elif moveType == JUMP_MOVE:
            moveType = move(JUMP_MOVE)
            showBoard()
        else:
            break
    print "######## GAME OVER ########\n\n"
    print "Thanks for Playing Checkers."
    print "Developed by: Nikhil Tarte"
```
